```python
#Print the number of rows for boy names and the number of rows for girl names.
```

```python
girls = list()
boys = list()
sex = df["sex"]

for x in sex:
    if x == 'F':
        girls.append(x)
    else:
        boys.append(x)

numgirls = len(girls)
numboys = len(boys)

print(numgirls)
print(numboys)
```

```python
print(len(df.loc[df['sex'] == 'F']))
print(len(df.loc[df['sex'] == 'M']))
```

```python
#The method Series.value_counts() computes a count of the unique values in a Series object. 
#Use this function to find out whether there are more distinct male or more distinct female names in the dataset.
```

```python

girlsunique = pd.Series(girls).value_counts()
boysunique = pd.Series(boys).value_counts()

print(girlsunique, boysunique)

#Er zijn meer unieke meisjesnamen dan jongensnamen
```

```python
#Find out how many distinct female names have a cumulative frequency higher than 100.
```

```python
female = list()
#freq = df['frequency']

for freq in df['frequency']:
    for sex in ('F', 'M'):
        if sex == 'F':
            if freq > 100:
                female.append(freq)

print(len(female))
```
